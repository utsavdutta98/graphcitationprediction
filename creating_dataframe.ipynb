```python
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# the encoding is important here otherwise you get a UnicodeDecodeError
with open('outputacm.txt', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
# now need to process lines
data = pd.DataFrame(columns = ['Index', 'Title', 'Authors', 'Year', 'Citations', 'Abstract'])
print(data)

i = 0
d = {}
for l in tqdm(range(int(len(lines)))): # full thing takes time to run
    if l == 0: # to handle the first line
        continue
    string = lines[l]

    if string == '\n':
        i = i + 1
        if 'Abstract' not in d.keys():
            d['Abstract'] = ''
        data = data.append(d, ignore_index = True)
        d = {} # reset
        continue
    
    if string.startswith('#*'):
        d['Title'] = string[2:-1] # start after #*, and end before \n at the end
    elif string.startswith('#@'):
        d['Authors'] = string[2:-1]
    elif string.startswith('#t'):
        d['Year'] = string[2:-1]
    elif string.startswith('#c'):
        d['Citations'] = string[2:-1]
    elif string.startswith('#!'):
        d['Abstract'] = string[2:-1]
    elif string.startswith('#index'):
        d['Index'] = string[6:-1]

print(i)
```

```python
data.to_csv('dataset_processed_full.csv', index = False)
```
